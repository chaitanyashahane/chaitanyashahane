### Removing the outliers for continuous variables

```python
columns=['HighestSpend','MonthlyAverageSpend','Mortgage']
```

```python
outliers = []
for c in columns:
    Q3 = np.percentile(df[c],75)
    Q1 = np.percentile(df[c],25)
    IQR = Q3 - Q1
    Threshold = IQR*1.5
    uf = Q3 + Threshold
    lf = Q1 - Threshold
    dfwoO = df.loc[(df[c] >= lf) & (df[c] <= uf)]
    mean= dfwoO[c].mean()
    df[c]=np.where(df[c]>uf,mean,df[c])
    df[c]=np.where(df[c]<lf,mean,df[c])
    #checking if there are any outliers remaining
    for i in df[c]:
        if i < lf or i > uf:
           outliers.append(i)
        
print(outliers)
```

### Although target variable is not balanced, lets check the performance of respective models 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
LR=LogisticRegression()
GN=GaussianNB()
KNN_model=KNeighborsClassifier(n_neighbors=5)
SVC = SVC(gamma=0.025, C=3)
```

```python
models = []
models.append(('LR', LR))
models.append(('NB', GN))
models.append(('KNN', KNN_model))
models.append(('SVC', SVC))
```

```python
# evaluate each model in turn
from sklearn import model_selection
results = []
names = []
scoring = 'accuracy'
for name, model in models:
	kfold = model_selection.KFold(n_splits=10, random_state=10,shuffle=True)
	cv = model_selection.cross_val_score(model,  x,y, cv=kfold, scoring=scoring)
	results.append(cv)
	names.append(name)
	msg = "%s: %f (%f)" % (name, cv.mean(), cv.std())
	print(msg)
```

#### It seems all model are performing nicely with unbalanced data but standard deviation is little more than it should be. SVC has great accuracy with reasonable standard deviation as compared to other modules

### Lets balance the target variable with upsampling the minority 

```python
from sklearn.utils import resample
df_majority = df_copy[df_copy.LoanOnCard == 0]
df_minority = df_copy[df_copy.LoanOnCard == 1]

df_min = resample(df_minority, replace = True, n_samples=4520,random_state=123)

df_balanced = pd.concat([df_majority, df_min])

df_balanced.LoanOnCard.value_counts()
```

```python
x = df_balanced.drop(['ID','CustomerSince','ZipCode', 'LoanOnCard'], axis=1)
y = df_balanced['LoanOnCard']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=10)
```

```python
y_train.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression

# Fit the model on train
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
model_score_train = model.score(x_train, y_train)
print(model_score_train)
model_score_test = model.score(x_test, y_test)
print(model_score_test)
```

```python
from sklearn import metrics
cm=metrics.confusion_matrix(y_test, y_predict, labels=[1, 0])

df_cm = pd.DataFrame(cm, index = [i for i in ["1","0"]],
                  columns = [i for i in ["Predict 1","Predict 0"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True,fmt='g')
```

```python
print("Classification Report for Logistic Regression")
print(metrics.classification_report(y_test, y_predict, labels=[1, 0]))
```

### Balancing the classes is resulting into really good result.
